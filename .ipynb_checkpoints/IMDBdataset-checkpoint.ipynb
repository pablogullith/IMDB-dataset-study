# IMDB TOP 2000 MOVIES DATASET - Pablo Gullith

```python
#Bibliotecas 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
import warnings
warnings.filterwarnings("ignore")
```

```python
#Carregando o dataset
df = pd.read_csv("imdb_top_2000_movies.csv")
df.head()
```

```python
df.info()
```

```python
df = df.dropna(subset="Gross").drop_duplicates(ignore_index=True)
df.isnull().sum() 
```

```python
#Função para transformar os valores to_numeric
import re

def to_numeric(x):
    pattern = r"[\d.]+"
    match = re.match(pattern, str(x))
    return float(match.group()) if match else None

df["Gross"] = df["Gross"].apply(to_numeric)
df["Votes"] = df["Votes"].apply(to_numeric)
df.head()
```

```python
#Top 10 filmes pelo valor bruto 
df = df.sort_values(by = 'Gross', ascending = False)
# Usando Seaborn para plotar o gráfico dos top 10 filmes por valor bruto com cores diferentes
plt.figure(figsize=(10, 6))
sns.barplot(x='Gross', y='Movie Name', data=df.head(10), palette='viridis')
plt.xlabel('Gross Value')
plt.ylabel('Movie Name')
plt.title('Top 10 Movies by Gross Value')
plt.show()
```

```python
#Top 10 filmes por nota no IMDB 
df = df.sort_values(by = 'IMDB Rating', ascending = False)
# Usando Seaborn para plotar o gráfico dos top 10 filmes pela nota do imdb
plt.figure(figsize=(10, 6))
sns.barplot(x='IMDB Rating', y='Movie Name', data=df.head(10), palette='viridis')
plt.xlabel('IMDB Rating')
plt.ylabel('Movie Name')
plt.title('Top 10 Movies by IMDB Rating')
plt.show()
```

```python
#Top 10 filmes por numero de votos 
df = df.sort_values(by = 'Votes', ascending = False)
# Usando Seaborn para plotar o gráfico dos top 10 filmes pela nota do imdb
plt.figure(figsize=(10, 6))
sns.barplot(x='Votes', y='Movie Name', data=df.head(10), palette='viridis')
plt.xlabel('Votes')
plt.ylabel('Movie Name')
plt.title('Top 10 Movies by Votes')
plt.show()
```

```python
#Top 10 diretores por valor bruto
# Ordenando o DataFrame pelos valores brutos em ordem decrescente
df = df.sort_values(by='Gross', ascending=False)

# Agrupando e somando por diretor, pegando os top 10
data = df.groupby('Director')['Gross'].sum().nlargest(10)

# Plotando o gráfico dos top 10 diretores por valor bruto
plt.figure(figsize=(9, 5))
sns.barplot(x=data.values, y=data.index, palette="viridis")
plt.title("Top 10 Directors By Gross")
plt.xlabel("Total Gross")
plt.ylabel("Director")
plt.tight_layout()
plt.show()
```

```python
#Top 10 cast por valor bruto
# Ordenando o DataFrame pelos valores brutos em ordem decrescente
df = df.sort_values(by='Gross', ascending=False)

# Agrupando e somando por cast, pegando os top 10
data = df.groupby('Cast')['Gross'].sum().nlargest(10)

# Plotando o gráfico dos top 10 cast por valor bruto
plt.figure(figsize=(9, 5))
sns.barplot(x=data.values, y=data.index, palette="viridis")
plt.title("Top 10 Cast By Gross")
plt.xlabel("Total Gross")
plt.ylabel("Cast")
plt.tight_layout()
plt.show()
```

```python
#Top 10 genero por valor bruto
# Ordenando o DataFrame pelos valores brutos em ordem decrescente
df = df.sort_values(by='Gross', ascending=False)

# Agrupando e somando por cast, pegando os top 10
data = df.groupby('Genre')['Gross'].sum().nlargest(10)

# Plotando o gráfico dos top 10 cast por valor bruto
plt.figure(figsize=(9, 5))
sns.barplot(x=data.values, y=data.index, palette="viridis")
plt.title("Top 10 Genre By Gross")
plt.xlabel("Total Gross")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()
```

```python
# Ordenando o DataFrame pelos valores de IMDB Rating em ordem decrescente
df = df.sort_values(by='IMDB Rating', ascending=False)

# Plotando o gráfico dos top 10 diretores por IMDB Rating sem as barras de erro
plt.figure(figsize=(9, 5))
sns.barplot(x='IMDB Rating', y='Director', data=df.head(10), palette="viridis", ci=None)
plt.title("Top 10 Directors By IMDB Rating")
plt.xlabel("IMDB Rating")
plt.ylabel("Director")
plt.tight_layout()
plt.show()
```

```python
#Top 10 cast by IMDB Rating 
# Ordenando o DataFrame pelos valores de IMDB Rating em ordem decrescente
df = df.sort_values(by='IMDB Rating', ascending=False)

# Plotando o gráfico dos top 10 diretores por IMDB Rating sem as barras de erro
plt.figure(figsize=(9, 5))
sns.barplot(x='IMDB Rating', y='Cast', data=df.head(10), palette="viridis", ci=None)
plt.title("Top 10 Cast By IMDB Rating")
plt.xlabel("IMDB Rating")
plt.ylabel("Cast")
plt.tight_layout()
plt.show()
```

```python
#Top 10 genre by IMDB Rating
# Ordenando o DataFrame pelos valores de IMDB Rating em ordem decrescente
df = df.sort_values(by='IMDB Rating', ascending=False)

# Plotando o gráfico dos top 10 diretores por IMDB Rating sem as barras de erro
plt.figure(figsize=(9, 5))
sns.barplot(x='IMDB Rating', y='Genre', data=df.head(10), palette="viridis", ci=None)
plt.title("Top 10 Genre By IMDB Rating")
plt.xlabel("IMDB Rating")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()
```

```python
#Top 10 movies and its Gross
df = df.sort_values(by='IMDB Rating', ascending=False)

# Plotando o gráfico dos top filmes por IMDB Rating e Gross Value
plt.figure(figsize=(10, 6))
sns.barplot(x='Gross', y='Movie Name', data=df.head(10), palette='viridis')
plt.xlabel('Gross Value')
plt.ylabel('Movie Name')
plt.title('Top 10 Movies by IMDB Rating and Gross Value')
plt.tight_layout()
plt.show()
```
